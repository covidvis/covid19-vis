```python
us_copy_grouped = us_copy_grouped.reset_index()
us_copy_grouped.head()
```

```python
capitals = data.us_state_capitals.url

# Taken from Altair documentation 

hover = alt.selection(type='single', on='mouseover', nearest=True,
                      fields=['lat', 'lon'])

base = alt.Chart(capitals).encode(
    longitude='lon:Q',
    latitude='lat:Q',
    tooltip=['state:N', 'Details (if any):N']
).transform_lookup(
    lookup='state',
    from_=alt.LookupData(data=us_copy_grouped, key='State',
                         fields=['Details (if any)'])
)

points = base.mark_point().encode(
    color=alt.value('blue'),
    size=alt.condition(~hover, alt.value(30), alt.value(100))
).add_selection(hover)

text = base.mark_text(dy=-5, align='right').encode(
    alt.Text('Details (if any):N'),
    opacity=alt.condition(~hover, alt.value(0), alt.value(1))
).transform_lookup(
    lookup='state',
    from_=alt.LookupData(data=us_copy_grouped, key='State',
                         fields=['Details (if any)'])
)

us_chart + points
```
